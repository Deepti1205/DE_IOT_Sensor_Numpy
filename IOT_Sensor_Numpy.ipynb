#**Transform**

###**Parse timestamps to Numpy datetime64**

```python
data_smart_home['Unix_Timestamp']
```

```python
timestamps = data_smart_home['Unix_Timestamp'].astype('datetime64[s]')
```

```python
timestamps  #np arrays are immutable so this will not be replaced inside the data_smart_home array
```

###**Combine with sensor measurements:**

```python
#sensor_cols = [col for col in data_smart_home.dtype.names if col !='Unix_Timestamp']
sensor_cols = [
    'Transaction_ID','Television', 'Dryer', 'Oven', 'Refrigerator', 'Microwave',
    'Line_Voltage', 'Voltage', 'Apparent_Power', 'Energy_Consumption_kWh','Hour_of_the_Day'
]
```

```python
values = np.vstack([data_smart_home[col].astype(float) for col in sensor_cols]).T
values
```

###**Define sampling windows (e.g., hourly):**

```python
timestamps
```

```python
timestamps[0]
```

```python
start = timestamps[0].astype('datetime64[h]')
start
```

```python
end = timestamps[-1].astype('datetime64[h]')
end
```

```python
hours = np.arange(start,end + 1,dtype='datetime64[h]')
hours
```

###**Compute hourly aggregates per sensor:**



```python
timestamps[:10]
```

```python
hours[:10]
```

**Elementwise BroadCasting**

```python
means = []
for i in range(10):
    mask = (timestamps >= hours[i]) & (timestamps < hours[i+1])
    print(mask)
    chunk = values[mask]
    print(chunk)
    #means.append(np.nanmean(chunk, axis=0))
```

```python
means = []
for i in range(len(hours)-1):
    mask = (timestamps >= hours[i]) & (timestamps < hours[i+1])
    #print(mask)
    chunk = values[mask]
    #print(chunk)
    means.append(np.nanmean(chunk, axis=0))
```

```python
means
```

```python
hourly_means = np.array(means)
```

```python
hourly_means
```

###**Detect and handle outliers using Z-score:**

```python
mean = np.nanmean(hourly_means, axis=0)
std = np.nanstd(hourly_means, axis=0)
z = (hourly_means - mean) / std
hourly_means[np.abs(z) > 3] = np.nan  # Mark as invalid
```

```python
hourly_means
```

#**LOAD**

###**Save cleaned data:**

```python
np.save("sensor_hourly_means.npy",hourly_means)
```
